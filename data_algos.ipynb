3.  Come up with a solution stated in plain english
4.  Implement solution and test using example inputs and fix bugs
5.  Analyze algorithm's complexity and identify inefficiencies, if any
6.  Apply right technique to overcome the inefficiencies. Repeat 3 - 6.

```python
# Binary Search
test = {
  'input': {
    'data': [50,40,30,21,17,12,6,3],
    'query': 21,
  },
    'output': 3
}

def locate_card_index(**data):
  
  # start with low, high points
  # we want the index, which starts at 0, which is one less than where we normally start
  low, high = 0, len(data['input']['data']) - 1
  while low <= high:
    mid = (low + high) // 2
    # find middle
    # print(mid)
    mid_value = data['input']['data'][mid]
    # print('low', low, 'high', high, 'mid', mid, 'mid value', mid_value)
    if mid_value == data['input']['query']: # if we hit our target
      print(mid_value)
      return mid
    elif mid_value < data['input']['query']: # if we overshoot our target
      high = mid - 1
      print('new high', high)
    elif mid_value > data['input']['query']: # if we undershoot our target
      low = mid + 1
      print('new low', low)
  return -1 # if the target is not found

locate_card_index(**test) == test['output']
```

```python
def locate_card(cards, query):
  lo, hi = 0, len(cards ) -1
  
  c = 0
  while lo <= hi:
    print('ran', c, 'times')
    mid = (lo + hi) // 2
    mid_number = cards[mid]
    c = c + 1
    
    if mid_number == query:
      return mid
    elif mid_number < query:
      hi = mid - 1
    elif mid_number > query:
      lo = mid + 1
  return -1

locate_card(list( range(10000000, 0, -1) ), 1)
# locate_card([1,3,5,8,10,12], 8)
```

```python
def count_linear_rotations(numbers):
  position = 0
  while position <= len(numbers):
    if position > 0 and numbers[position] < numbers[position - 1]:
      return position
    position +=1
  return 0
```

```python
class UserDatabase:
  def __init__(self) -> None:
    self.users = []
  def insert(self, user):
    i = 0
    while i < len(self.users):
      if self.users[i].username > user.username:
        break
      i += 1
    self.users.insert(i, user)

  def find(self, username):
    for user in self.users:
      if user.username == username:
        return user
      else: 
        return -1
  def update(self, user):
      target_user = self.find(user.username)
      target_user.name, target_user.email = user.name, user.email
      
  def list_all(self):
    return self.users
```

N is number of nodes in a tree
N = 1 + 2^1 + 2^2 + 2^3 + ... + 2^(k-1)
N + 1 = 1 + 1 + 2^1 + 2^2 + 2^3 + ... + 2^(k-1)
N + 1 = 2^1 + 2^1 + 2^2 + 2^3 + ... + 2^(k-1)
N + 1 = 2^2 + 2^2 + 2^3 + ... + 2^(k-1)
...
N + 1 = 2^(k-1) + 2^(k-1)
N + 1 = 2^k
k = log(N + 1) <= log(N) + 1

Thus, to store N records we require a balanced BST of height <= log(N) + 1

```python
class TreeNode:
   def __init__(self, data, left = None, right = None):
      self.data = data
      self.left = left
      self.right = right

def insert(temp,data):
   que = []
   que.append(temp)
   while (len(que)):
      temp = que[0]
      que.pop(0)
      if (not temp.left):
         temp.left = TreeNode(data)
         break
      else:
         que.append(temp.left)
      if (not temp.right):
         temp.right = TreeNode(data)
         break
      else:
         que.append(temp.right)
         
def make_tree(elements):
   Tree = TreeNode(elements[0])
   for element in elements[1:]:
      insert(Tree, element)
   return Tree
  
class Solution(object):
   def inorderTraversal(self, root):
      res, stack = [], []
      current = root
      while True:
         while current:
            stack.append(current)
            current = current.left
         if len(stack) == 0:
            return res
         node = stack[-1]
         stack.pop(len(stack)-1)
         if node.data != None:
            res.append(node.data)
         current = node.right
      return res
ob1 = Solution()
root = make_tree([10,5,15,2,7,None,20])
print(ob1.inorderTraversal(root))
# input [10,5,15,2,7,null,20]
# output [2,5,7,10,15,20]

```

```python
# O(n)
def insert_sort(A):
  for j in range(1, len(A)):
    key = A[j]
    i = j - 1
    while i >= 0 and A[i] > key:
      A[i + 1] = A[i]
      i -= 1
    A[i + 1] = key
  return A
      
```

```python
def merge(left, right):
  merged = []
  left_index, right_index = 0,0
  
  while left_index < len(left) and right_index < len(right):
    if left[left_index] <= right[right_index]:
      merged.append(left[left_index])
      left_index +=1
    else:
      merged.append(right[right_index])
      right_index +=1
      
  remaining_left = left[left_index:]
  remaining_right = right[right_index:]
  return merged + remaining_left + remaining_right
      
```

```python
# Time O(n log n)
# Space O(n)
# Total comparisons n * h where h is tree hight and n is total leaf nodes
# Since lowest level sublist is size 1 and 1 sublist is compared with another of size 1
# 2^(h-1) = n -> h = log n + 1

def merge_sort(mnums):
  # print(mnums)
  if len(mnums) <= 1:
    return mnums
  mid = len(mnums) // 2
  left, right = mnums[:mid], mnums[mid:]

  left_itterator, right_itterator, current_itterator = 0,0,0
  
  left_sorted, right_sorted = merge_sort(left), merge_sort(right)
  left_size = len(left_sorted)
  right_size = len(right_sorted)
  while left_itterator < left_size and right_itterator < right_size:
    if left[left_itterator] < right[right_itterator]:
      mnums[current_itterator] = left[left_itterator]
      left_itterator += 1
      current_itterator += 1
    else:
      mnums[current_itterator] = right[right_itterator]
      right_itterator += 1
      current_itterator += 1
      
    while left_itterator < left_size:
      mnums[current_itterator] = left[left_itterator]
      left_itterator += 1
      current_itterator += 1
 
    while right_itterator < right_size:
      mnums[current_itterator]=right[right_itterator]
      right_itterator += 1
      current_itterator += 1
  sorted_nums = merge(left_sorted, right_sorted)
  return sorted_nums

print( merge_sort(get_random()))
```
